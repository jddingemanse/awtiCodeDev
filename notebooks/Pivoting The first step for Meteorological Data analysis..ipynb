# This code makes NMI Data ready for missing data analysis this process is the basic step for any meteorological data processing if you are interested in the first step of meteorological data analysis pls flow all the steps below and let me know if there is error. 

```python
# Importing the requared packeges
```

```python
import pandas as pd
import numpy as np
```

```python
# Importing the requared Data to process. you can download 'NMI.xlsx' from awti code data folder I will put it their.
# if you are unabel to get it let me know.
```

```python
Data =pd.read_excel('NMI.xlsx')
Data
```

```python
# now let separet the months as the months starts from the forth colomun from our data let us do this
```

```python
monthnames = Data.columns[4:]
monthnames
```

```python
# replacing month name by month number will help as alot for data analysis so as to do so let us use this code
```

```python
monthnumbers = np.arange(1,13)
monthnumbers
```

```python
#let us creat a data frame that contain a columen name year,month,day,rainfall,stations and elements
```

```python
dataLong = Data.melt(id_vars=['Stations','Elements','Date','Year'],var_name='Month',value_name='Rainfall')
dataLong = dataLong.get(['Year','Month','Date','Rainfall','Stations','Elements']).set_axis(['year','month','day','Rainfall','Stations','Elements'],axis=1)
```

```python
dataLong
```

```python
# now let us replace month name by month number
```

```python
dataLong.loc[:,'month']=dataLong.month.replace(monthnames,monthnumbers)
dataLong
```

```python
#now let us prepare time data frame that contains  a colomun of year,month and day this will help us for timestamp
```

```python
timedf = dataLong.get(['year','month','day'])
timedf
```

```python
# now let us make timestamp
```

```python
dataLong['timestamp'] = pd.to_datetime(timedf,errors='coerce')
dataLong.sort_values(by=['year','month','day'],inplace=True)
dataLong
```

```python
# so as to avoid repetation let us drop some colomen i.e year,month and day because they are already included in timestamp
# and let us save it as A
```

```python
A= dataLong.drop(columns=['year','month','day'])
A.sort_values(by=['timestamp','Stations','Rainfall'],inplace=True)
A.dropna(subset=['timestamp'],inplace=True)
A
```

```python
# now let us creat commen startind and ending date for all station and let us save it as A_1992_2008
```

```python
startyear=pd.to_datetime('1992')
endyear=pd.to_datetime('2009')
A_1992_2008=A.loc[(A.timestamp >= startyear) & (A.timestamp <= endyear)]
A_1992_2008
```

```python
# Now let us use pivoting so as to prepare our data for missing data analysis and let as save it as pivoted
# we can also export in to csv as 'final.csv' indicated below.
```

```python
index=['timestamp']
pivoted = A_1992_2008.pivot(index= index, columns='Stations', values='Rainfall')
pivoted
pivoted.to_csv('final.csv')
```
