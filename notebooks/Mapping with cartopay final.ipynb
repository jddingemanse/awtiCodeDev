```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker
import cartopy
from cartopy import crs as ccrs, feature as cfeature
```

```python
# ploting
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.subplot(1, 1, 1, projection=ccrs.PlateCarree(central_longitude=10))
ax.set_title("A Geo-referenced subplot, Plate Carree projection");
ax.coastlines(linewidth=1.0, linestyle='solid', color='black')
```

```python
# adding Borders
```

```python
ax.add_feature(cfeature.BORDERS, linewidth=1.0,linestyle='solid', edgecolor='black')
fig
```

```python
#adding feature(land,lakes and ocean)
```

```python
ax.add_feature(cfeature.LAND, edgecolor='coral', facecolor="#ddaa66", zorder=0)
ax.add_feature(cfeature.LAKES, edgecolor='g', linewidth=0.2, facecolor='g')
ax.add_feature(cfeature.OCEAN,facecolor="#7777ff")
fig
```

```python
#adding Grid lines
```

```python
gl = ax.gridlines(crs=ccrs.PlateCarree(), linewidth=1, color='black', alpha=0.5)
gl.ylocator = mticker.FixedLocator(np.arange(-90,90,20))
gl.xlocator = mticker.FixedLocator(np.arange(-180, 180, 20))
gl = ax.gridlines(draw_labels=True, linewidth=2, color='gray', alpha=0.5, linestyle='--')
fig
```

```python
#adding rivers
```

```python
rivers_50m = cfeature.NaturalEarthFeature('physical', 'rivers_lake_centerlines', '10m')
ax.add_feature(rivers_50m, facecolor='None', edgecolor='g')
fig
```

```python
#ploting with different projections 
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Orthographic(25))
ax.add_feature(cfeature.LAND, edgecolor='r', facecolor="none", zorder=10)
ax.add_feature(cfeature.BORDERS, linewidth=1.0,linestyle='solid', edgecolor='k')
ax.add_feature(cfeature.OCEAN,facecolor='b')
ax.gridlines()
ax.set_title(" Orthographic_Projection", fontsize=20)
#ax.set_title("A Geo-referenced subplot, Plate Carree projection");
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.InterruptedGoodeHomolosine(0))
ax.add_feature(cfeature.LAND, edgecolor='r', facecolor="none", zorder=10)
ax.add_feature(cfeature.BORDERS, linewidth=1.0,linestyle='solid', edgecolor='k')
ax.add_feature(cfeature.OCEAN,facecolor='b')
ax.gridlines()
ax.set_title("InterruptedGoodeHomolosine_Projection", fontsize=20)
```

```python
# maping for africa
```

```python
fig = plt.figure(figsize=(11, 11))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_extent([-20, 60, -40, 45], crs=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, edgecolor='coral', facecolor="#ddaa66", zorder=0)
ax.add_feature(cfeature.LAKES, edgecolor='g', linewidth=0.2, facecolor='g')
ax.add_feature(cfeature.OCEAN,facecolor="b")
ax.add_feature(cfeature.BORDERS, linestyle='solid')
ax.set_title("Africa", fontsize=30);
plt.show()
```

```python
#mapping for Ethiopia
```

```python
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([32, 49, 3, 15], crs=ccrs.PlateCarree())
ax.add_feature(cfeature.LAND, edgecolor='g', facecolor="#ddaa66")
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.COASTLINE)
ax.gridlines(xlocs=np.arange(20,70,3), draw_labels=True, crs=ccrs.PlateCarree())
ax.set_title("A Geo-referenced subplot, Ethiopia");
```
