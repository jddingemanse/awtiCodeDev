```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import netCDF4 as nc
import io
import numpy as np
from datetime import datetime,timedelta
import metpy
from metpy.units import units
from siphon.catalog import TDSCatalog
import xarray as xr
from xarray.backends import NetCDF4DataStore
import pandas as pd
from netCDF4 import Dataset
from scipy.io import netcdf
```

```python
from netCDF4 import Dataset as dataset

# Retreiving dataset
Dta_URL = ('http://thredds.ucar.edu/thredds/catalog/grib/NCEP/GFS/'
              'Global_0p25deg/catalog.xml?dataset=grib/NCEP/GFS/Global_0p25deg/Best')

best_gfs = TDSCatalog(Dta_URL)
print(list(best_gfs.datasets))

best_ds = best_gfs.datasets[0]


ncss = best_ds.subset()
query = ncss.query()

query.lonlat_box(north=15, south=3, east=49, west=32.5).time_range(datetime.utcnow(), 
                                                                   datetime.utcnow() + timedelta(days=1))
query.accept('netcdf4')
query.variables('Precipitation_rate_surface')
data = ncss.get_data_raw(query)

gfs = xr.open_dataset(io.BytesIO(data))
lat = gfs.variables['latitude'][:]
lon = gfs.variables['longitude'][:]
time = gfs.variables['time1'][:]

ppt_day1 = gfs.variables['Precipitation_rate_surface'][:] *86400
ppt_day1_sum = np.sum(ppt_day1, axis = 0)

fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
cs = m.contourf(x, y, ppt_day1_sum, levels = 100, cmap=plt.cm.jet)
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries(linewidth=1, linestyle='solid', color='k' ) 
m.drawmeridians(range(33, 48, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[0, 0, 0, 1])
m.drawparallels(range(3, 15, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[1, 0, 0, 0])
plt.ylabel("Latitude", fontsize=15, labelpad=35)
plt.xlabel("Longitude", fontsize=15, labelpad=20)
cbar = m.colorbar(cs, location='right', pad="3%")
cbar.set_label('ppt(mm)', fontsize=13)
plt.title('Precipitation_rate_surface contour map day1', fontsize=15)
plt.show()

query.lonlat_box(north=15, south=3, east=49, west=32.5).time_range(datetime.utcnow() + timedelta(days=1), 
                                                                   datetime.utcnow() + timedelta(days=2))
query.accept('netcdf4')
query.variables('Precipitation_rate_surface')
data = ncss.get_data_raw(query)
gfs = xr.open_dataset(io.BytesIO(data))
lat = gfs.variables['latitude'][:]
lon = gfs.variables['longitude'][:]
time = gfs.variables['time1'][:]
ppt_day2 = gfs.variables['Precipitation_rate_surface'][:] *86400
ppt_day2_sum = np.sum(ppt_day2, axis = 0)

fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
cs = m.contourf(x, y, ppt_day2_sum, levels = 100, cmap=plt.cm.jet)
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries(linewidth=1, linestyle='solid', color='k' ) 
m.drawmeridians(range(33, 48, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[0, 0, 0, 1])
m.drawparallels(range(3, 15, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[1, 0, 0, 0])
plt.ylabel("Latitude", fontsize=15, labelpad=35)
plt.xlabel("Longitude", fontsize=15, labelpad=20)
cbar = m.colorbar(cs, location='right', pad="3%")
cbar.set_label('ppt(mm)', fontsize=13)
plt.title('Precipitation_rate_surface contour map day2', fontsize=15)
plt.show()

query.lonlat_box(north=15, south=3, east=49, west=32.5).time_range(datetime.utcnow() + timedelta(days=2), 
                                                                   datetime.utcnow() + timedelta(days=3))
query.accept('netcdf4')
query.variables('Precipitation_rate_surface')
data = ncss.get_data_raw(query)
gfs = xr.open_dataset(io.BytesIO(data))
lat = gfs.variables['latitude'][:]
lon = gfs.variables['longitude'][:]
time = gfs.variables['time1'][:]
ppt_day3 = gfs.variables['Precipitation_rate_surface'][:] *86400
ppt_day3_sum = np.sum(ppt_day3, axis = 0)


fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
fig = plt.figure(num=None, figsize=(7, 7) ) 
m = Basemap(projection='cyl', llcrnrlon=32.5, llcrnrlat=3, urcrnrlon=49, urcrnrlat=15, resolution='l')
x, y = m(*np.meshgrid(lon,lat))
cs = m.contourf(x, y ,ppt_day3_sum, levels = 100, cmap=plt.cm.jet)
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries(linewidth=1, linestyle='solid', color='k' ) 
m.drawmeridians(range(33, 48, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[0, 0, 0, 1])
m.drawparallels(range(3, 15, 2), color='k', linewidth=1.0, dashes=[4, 4], labels=[1, 0, 0, 0])
plt.ylabel("Latitude", fontsize=15, labelpad=35)
plt.xlabel("Longitude", fontsize=15, labelpad=20)
cbar = m.colorbar(cs, location='right', pad="3%")
cbar.set_label('ppt(mm)', fontsize=13)
plt.title('Precipitation_rate_surface contour map day3', fontsize=15)
plt.show()
```
